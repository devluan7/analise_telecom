```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from cycler import cycler
import numpy as np


df_telecom = pd.read_csv("../database/clientes_telecom.csv")
```

```python
#Farei uma an√°lise da qauntidade pessoas que assinam cada plano (Anual, Bianual e Mensal)
df_telecom_contrato = df_telecom["contrato"].value_counts()
```

```python
#Gr√°fico da an√°lise
fig, ax = plt.subplots(figsize = (6, 6))
#cor dos gr√°ficos
cores = plt.get_cmap("Accent").colors
ciclo_cores = cycler("color", cores)
plt.rc("axes", prop_cycle=ciclo_cores)

#Configurando a fatia destacada
maior_indice = np.argmax(df_telecom_contrato.values)
destaque = [0.1 if i == maior_indice else 0 for i in range(len(df_telecom_contrato.values))]

ax.pie(df_telecom_contrato.values, labels = df_telecom_contrato.index, autopct ="%1.0f%%", pctdistance = 0.3, explode = destaque)
plt.suptitle("Porcentagem de assinantes por plano", fontsize = 15)

fig.legend(loc = "lower center", prop={'size': 13})

plt.savefig("../outputs/pct_assinantes_plano.jpg", bbox_inches='tight')
plt.show()
```
